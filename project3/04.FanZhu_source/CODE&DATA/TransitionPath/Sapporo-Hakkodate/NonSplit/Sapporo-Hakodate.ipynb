```python
score = np.max(score_data.values[[0,2,4]],axis=0)
N = score.shape[0]
pd.DataFrame(score.reshape((1,-1)), columns=city, index=['score'])
```

## Transition Matrix

```python
tran_mat = np.matmul(adj_mat, np.diag(score))
deg = np.sum(tran_mat,axis=1)
tran_mat = np.matmul(np.diag(1/deg), tran_mat)
```

```python
df = pd.DataFrame(tran_mat, index=city, columns=city)
print(df)
df.to_csv('Tran_mat.csv')
```

## Stationary Distribution

Solve a linear equation $\pi * P = \pi$. Here we use package ``discreteMarkovChain``

```python
mc = markovChain(P=tran_mat)
```

```python
mc.computePi('linear')
pd.DataFrame(mc.pi.reshape((1,-1)), columns=city, index=['Prob'])
```

## Committor function

Solve a (discrete) Dirichlet boundary problem for committor function, here we should let the boundary be the first col and last col.

```python
a1 = np.zeros(N)
a1[0] = 1
a2 = np.zeros(N)
a2[-1] = 1
a = np.vstack((tran_mat[1:-1,0], (tran_mat[1:-1,1:-1]-np.identity(N-2)).T, tran_mat[1:-1,-1])).T
a = np.vstack((a1,a,a2))
```

```python
b = np.zeros(N)
b[0] = 0
b[-1] = 1
```

```python
q = np.linalg.solve(a,b)
pd.DataFrame(q.reshape(1,-1), columns=city, index=['commitor'])
```

## Effective Flux 

Solve the effective flux on each edge (x,y)
Firstly, calculate reactive current, which describes the ``reactive flow passing by x and y``.
Then, effective flux is basically the max(J(xy)-J(yx),0)

```python
J_dir = np.zeros((N,N))
for x in range(N):
    for y in range(N):
        J_dir[x,y] = mc.pi[x] * (1-q[x]) * tran_mat[x,y] * q[y]
```

```python
J_eff = np.maximum(J_dir - J_dir.T, 0)
print(J_eff[0])
np.max(J_eff)
```

```python
u = J_eff[0]
np.percentile(a=u,q=66)
```

```python
np.percentile(J_eff[J_eff!=0],q=66)
```

## Transition Current of a Node

Compute the transition flux through each node x$\in$V.

```python
flux = np.zeros(N)
flux[0] = np.sum(J_eff[0])
flux[-1] = np.sum(J_eff[:,-1])
for x in np.arange(1,N-1):
    flux[x] = np.sum(J_eff[x,])
    #print(J_eff[x,].sum() - J_eff[:,x].sum())
print(flux)
```

```python
pd.DataFrame( np.reshape( flux / np.sum(flux[1:-1]), (1,-1)), index=['Current'], columns=city)
```

## Graph Force Layout

```python
nodes = []
for index, name in enumerate(city):
    if index == 0 or index == N-1:
        nodes.append({
            'id':name,
            'group':'red',
            'radius':np.mean(flux[1:-1])/2+np.max(flux[1:-1])/2
        })
    else:
        nodes.append({
            'id':name,
            'group':'blue',
            'radius':flux[index]
        })
```

```python
nodes[5]['group'] = 'pink'
nodes[-5]['group'] = 'pink'
nodes
```

```python
mean_0 = np.mean(J_eff[0][J_eff[0]!=0])
mean_n = np.mean(J_eff[:,-1][J_eff[:,-1]!=0])
mean = np.percentile(J_eff[1:-1,1:-1][J_eff[1:-1,1:-1]!=0], q=65)
J_eff_copy = J_eff.copy()
J_eff_copy[0] *= mean/mean_0
J_eff_copy[:,-1] *= mean/mean_n
```

```python
cut_thresh = np.percentile(J_eff[1:-1,1:-1][J_eff[1:-1,1:-1]!=0],q=75) # stress 1/5
```

```python
def f(index1, index2, dis, cut_thresh=cut_thresh):
    if index1 == 0 or index1 == N-1 or index2 == 0 or index2==N-1:
        return 'red'
    elif dis > cut_thresh:
        return 'green'
    else:
        return 'gray'
```

```python
links = []
for index1, a in enumerate(adj_mat):
    for index2, b in enumerate(a):
        if index1>index2: 
            if J_dir[index1,index2] > J_dir[index2,index1]:
                links.append({
                    'source':city[index1],
                    'target':city[index2],
                    'value':J_eff_copy[index1,index2],
                    'group': f(index1, index2, J_eff[index1, index2])
                })
            elif J_dir[index2,index1] > J_dir[index1,index2]:
                links.append({
                    'source':city[index2],
                    'target':city[index1],
                    'value':J_eff_copy[index2,index1],
                    'group': f(index1, index2, J_eff[index2, index1])
                })
```
